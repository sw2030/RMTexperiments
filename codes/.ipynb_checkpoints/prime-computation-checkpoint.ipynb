```julia
using Plots, LinearAlgebra, ForwardDiff, Distributions, SpecialFunctions 
import GaussQuadrature.legendre
```

```julia
∂ = ForwardDiff.derivative;
function GL(m; a=0, b=1) 
    # m points Gauss-Legendre quadrature for [a, b]
    xi, wi = legendre(m)
    return map( t->((a+b)/2 + t*(b-a)/2), xi ), (b-a) * wi / 2
end
function K_quadrature(K, s; m=50) 
    # Creates K matrix according to m quadrature points
    # Method proposed in (Bornemann, 2010)
    x, w = GL(m; a=0, b=s)
    w_sqrt = sqrt.(w)
    return (w_sqrt*w_sqrt') .* map(t->K(t...), [(xi,xj) for xi in x, xj in x]) 
end
```

# 1. Sine kernel and $\det(I-K|_{L^2(0,s)}) = g(s)$ ($=E_2(0;s)$)

```julia
## Derived kernels of the Sine Kernel
K_sinc(x, y) = x==y ? 1.0 : sin( π * (x - y) ) / ( π * (x - y) ) 
K0_sinc(x) = sin( π*x )/ (π*x)                                           # K⁽⁰⁾
Kp_sinc(x, y) = K_sinc(x, y) - K0_sinc(x) * K0_sinc(y)                   # K⁽ˢ⁾ 
Kp_sinc_B(x, y) = cos(π*(x-y))   # Derived kernel [Bornemann 2022], Theorem 3.1  
g(s,m=50) = det(I - K_quadrature(K_sinc, s; m=m))
g(0.1) # E2(0, 0.1) = 0.90002 72717 98259 
```

## (a) approximate $\frac{d}{ds}g(s) = f(s)$, different number of quadrature points

```julia
f_A(s,m) = ∂(t->g(t,m), s)                            # Autodiff of E₂(s)
f_B(s,m) = -g(s,m)*tr((I-K_quadrature(K_sinc,s;m=m))
                    \K_quadrature(Kp_sinc_B,s;m=m))/s   # (Bornemann 2022), Theorem 3.1
f_C(s,m) = -det(I - K_quadrature(Kp_sinc, s;m=m))      # conditional DPP kernel

s = 0.5
REF_VAL = mean(map(t->f_A(s,t), 100:100:1000))
println("s=$s and reference value f(s)=$REF_VAL")
```

```julia
ms = 5:20 # number of different quadrature points
err = abs.([f_A.(s,ms) f_B.(s,ms) f_C.(s,ms)] .- REF_VAL);
plot(yaxis=:log, xlabel="# quad points", ylabel="error", title="Error of computing f(s)")
plot!(ms, err[:, 1], label="Autodiff")
plot!(ms, err[:, 2], label="derived kernel")
plot!(ms, err[:, 3], label="conditional kernel")
```

## (b) approximate $v(s) = s\frac{f(s)}{g(s)}$

```julia
v_A(s,m) = s*f_A(s,m)/g(s,m) 
v_B(s,m) = -tr( (I - K_quadrature(K_sinc,s;m=m)) \ K_quadrature(Kp_sinc_B,s;m=m) )
v_C(s,m) = s*f_C(s,m)/g(s,m) 

s = 2.5
REF_VAL = mean(map(t->(s * f_A(s,t) / g(s,t)), 100:100:1000))
println("s=$s and reference value v(s)=$REF_VAL")
```

```julia
ms = 5:20
err = abs.([v_A.(s,ms) v_B.(s,ms) v_C.(s,ms)] .- REF_VAL);
plot(yaxis=:log, xlabel="# quad points", ylabel="error", title="Error of computing v(s)")
plot!(ms, err[:, 1], label="Autodiff")
plot!(ms, err[:, 2], label="derived kernel")
plot!(ms, err[:, 3], label="conditional kernel")
```

# 2. Bessel kernel and $\det(I-K|_{L^2(0,s)}) = g_h(s)$ 

```julia
## Bessel kernel and derived kernels
Ja = besselj
K_bessel(α,x,y) = x==y ? ( Ja(α, √x)^2 - Ja(α+1, √x) * Ja(α-1, √x) ) / 4 :
                                 ( Ja(α+1, √x) * √x * Ja(α, √y) - √y * Ja(α+1, √y) * Ja(α, √x) ) / ( 2 * (x - y) )
Kp_bessel(α,x,y,s) = K_bessel(α, x, y) - K_bessel(α, s, x) * K_bessel(α, s, y) / K_bessel(α, s, s)
Kp_bessel_B(α,x,y) = Ja(α, √x) * Ja(α, √y) / 4
g_h(α,s,m=50) = det(I - K_quadrature((x,y)->K_bessel(α,x,y),s;m=m)) # h for hard-edge
```

## (a) Approximate $\frac{d}{ds}g_h(s) = f_h(s)$

```julia
f_A_h(α,s,m) =  ∂(t->g_h(α,t,m),s)
f_B_h(α,s,m) = -g_h(α,s,m)*tr((I-K_quadrature((x,y)->K_bessel(α,x,y),s;m=m))
                                    \K_quadrature((x,y)->Kp_bessel_B(α,x,y),s;m=m))/s
f_C_h(α,s,m) = -det(I-K_quadrature((x,y)->Kp_bessel(α,x,y,s),s;m=m)) * K_bessel(α,s,s)

s = 100
α = 10
REF_VAL = mean(map(t->f_A_h(α,s,t), 100:100:1000))
println("s=$s, α=$α and reference value f(s)=$REF_VAL")
```

```julia
ms = 5:20
err = abs.([f_A_h.(α,s,ms) f_B_h.(α,s,ms) f_C_h.(α,s,ms)] .- REF_VAL)
plot(yaxis=:log, xlabel="# quad points", ylabel="error", title="Error of computing fₕ(s)")
plot!(ms, err[:, 1], label="Autodiff")
plot!(ms, err[:, 2], label="derived kernel")
plot!(ms, err[:, 3], label="conditional kernel")
```

## (b) Approximate $v_h(s) = s\frac{f_h(s)}{g_h(s)}$ for bessel

```julia
v_A_h(α,s,m) = s*f_A_h(α,s,m)/g_h(α,s,m) 
v_B_h(α,s,m) = -tr((I-K_quadrature((x,y)->K_bessel(α,x,y),s;m=m))\K_quadrature((x,y)->Kp_bessel_B(α,x,y),s;m=m))
v_C_h(α,s,m) = s*f_C_h(α,s,m)/g_h(α,s,m) 

s = 100
α = 10
REF_VAL = mean(map(t->v_A_h(α,s,t), 100:100:1000))
println("s=$s, α=$α and reference value f(s)=$REF_VAL")
```

```julia
ms = 5:20
err = abs.([v_A_h.(α,s,ms) v_B_h.(α,s,ms) v_C_h.(α,s,ms)] .- REF_VAL)
plot(yaxis=:log, xlabel="# quad points", ylabel="error", title="Error of computing vₕ(s)")
plot!(ms, err[:, 1], label="Autodiff")
plot!(ms, err[:, 2], label="derived kernel")
plot!(ms, err[:, 3], label="conditional kernel")
```

# 3. Airy kernel and $\det(I-K|_{L^2(s,\infty)}) = g_s(s)$ 

```julia
## Airy kernel and derived kernels
Ai = airyai
Aip = airyaiprime
ϕ(x,s) = s+10*tan(π*x/2) # See Bornemann 2010 pdf page 29
ϕp(x) = 5π*sec(π*x/2)^2
K_Ai(x, y) = x==y ? (Aip(x))^2 - x * (Ai(x))^2 : (Ai(x) * Aip(y) - Ai(y) * Aip(x)) / (x - y)
K_Ai_ϕ(s,x,y) = sqrt( ϕp(x) * ϕp(y) ) * K_Ai( ϕ(x, s), ϕ(y, s) )
Kp_Ai(s,x,y) = K_Ai(x, y) - K_Ai(x, s)*K_Ai(s, y)/K_Ai(s, s)
Kp_Ai_ϕ(s,x,y) = sqrt( ϕp(x) * ϕp(y) ) * Kp_Ai(s,ϕ(x, s),ϕ(y, s))
Aivec(s,m) = Ai.(ϕ.(GL(m)[1],s)) .* sqrt.(GL(m)[2] .* ϕp.(GL(m)[1]))
g_s(s,m=50) = det(I-K_quadrature((x,y)->K_Ai_ϕ(s,x,y),1;m=m)) #
```

## (a) Approximate $\frac{d}{ds}g_s(s) = f_s(s)$

```julia
f_A_s(s,m) = ∂(t->g_s(t,m),s)
f_B_s(s,m) = g_s(s,m)*Aivec(s,m)'*((I-K_quadrature((x,y)->K_Ai_ϕ(s,x,y),1;m=m))\Aivec(s,m))
f_C_s(s,m) = det(I-K_quadrature((x,y)->Kp_Ai_ϕ(s,x,y),1;m=m)) * K_Ai(s,s)

s = -0.5
REF_VAL = mean(map(t->f_C_s(s,t), 100:20:300))
println("s=$s and reference value f(s)=$REF_VAL")
```

```julia
ms = 5:20
err = abs.([f_A_s.(s,ms) f_B_s.(s,ms) f_C_s.(s,ms)] .- REF_VAL)
plot(yaxis=:log, xlabel="# quad points", ylabel="error", title="Error of computing fₛ(s)")
plot!(ms, err[:, 1], label="Autodiff")
plot!(ms, err[:, 2], label="derived kernel")
plot!(ms, err[:, 3], label="conditional kernel")
```

## (b) Approximate $v_s(s) = s\frac{f_s(s)}{g_s(s)}$ for Airy kernel

```julia
v_A_s(s,m) = s*f_A_s(s,m)/g_s(s,m) 
v_B_s(s,m) = s*Aivec(s,m)'*((I-K_quadrature((x,y)->K_Ai_ϕ(s,x,y),1;m=m))\Aivec(s,m))
v_C_s(s,m) = s*f_C_s(s,m)/g_s(s,m) 

s = -2.5
REF_VAL = mean(map(t->v_A_s(s,t), 100:100:300))
println("s=$s and reference value f(s)=$REF_VAL")
```

```julia
ms = 5:20
err = abs.([v_A_s.(s,ms) v_B_s.(s,ms) v_C_s.(s,ms)] .- REF_VAL)
plot(yaxis=:log, xlabel="# quad points", ylabel="error", title="Error of computing vₛ(s)")
plot!(ms, err[:, 1], label="Autodiff")
plot!(ms, err[:, 2], label="derived kernel")
plot!(ms, err[:, 3], label="conditional kernel")
```
